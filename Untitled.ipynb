```python
import pandas as pd
import os 
from datetime import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#read data 
path = "/Users/savannahestridge/Desktop/gjny-exercise-main/nyc_admissions.csv"
df = pd.read_csv(path)

path2 = "/Users/savannahestridge/Desktop/gjny-exercise-main/albany_admissions.csv"
df2 = pd.read_csv(path2)
```

```python
# filter uneccesary data 
#The number of race == unknown & ethnicity == hispanic was too large for me to remove, 
#converted to race == Unknown (Hispanic)

df = df.drop(columns=['ethnicity'])
values = ['Other', 'Parole', "Unknown"]
df = df[df.admission_status.isin(values) == False]

df2 = df2.drop(df2[(df2['race'] == "Unknown") & (df2['ethnicity'] == "Non-Hispanic")].index)
df2['race'].replace('Unknown','Unknown (Hispanic)', inplace=True)
df2 = df2.drop(columns=['ethnicity'])
df2 = df2[df2.admission_status.isin(values) == False]

#Change to appropriate data type
df[['admission_date','release_date']] = df[['admission_date','release_date']].apply(pd.to_datetime)
df = df.sort_values(by=['admission_date'])

df2[['admission_date','release_date']] = df2[['admission_date','release_date']].apply(pd.to_datetime)
df2 = df2.sort_values(by=['admission_date'])
```

```python
#2019 vs 2020
nyc_19 = df.loc[(df['admission_date'] >= '2019-01-01')
                     & (df['admission_date'] <= '2019-12-31')
                     & (df["bail_eligible"] == "Eligible")]

nyc_20 = df.loc[(df['admission_date'] >= '2020-01-01')
                     & (df['admission_date'] <= '2020-12-31')
                     & (df["bail_eligible"] == "Eligible")]

alb_19 = df2.loc[(df2['admission_date'] >= '2019-01-01')
                     & (df['admission_date'] <= '2019-12-31')
                     & (df["bail_eligible"] == "Eligible")]

alb_20 = df2.loc[(df2['admission_date'] >= '2020-01-01')
                     & (df['admission_date'] <= '2020-12-31')
                     & (df["bail_eligible"] == "Eligible")]
```

# What is the makeup of charges (violent/non-violent) before and after the bail reform.

```python
#remove unnecessary data (This ends up removing all native american data because of null values)
nyc_sev = df.dropna(subset = ['severity'])
nyc_sev = nyc_sev.drop(['admission_status', 'charge', 'release_date', 'bail_eligible'], axis = 1)

alb_sev = df2.dropna(subset = ['severity'])
alb_sev = alb_sev.drop(['admission_status', 'charge', 'release_date', 'bail_eligible'], axis = 1)

value = ["Unknown"]
nyc_sev = nyc_sev[nyc_sev.race.isin(value) == False]
alb_sev = alb_sev[alb_sev.race.isin(value) == False]

#Filter for date 
nyc_sev_19 =nyc_sev.loc[(df['admission_date'] >= '2019-01-01')
                     & (df['admission_date'] <= '2019-12-31')]                   

nyc_sev_20 = nyc_sev.loc[(df['admission_date'] >= '2020-01-01')
                     & (df['admission_date'] <= '2020-12-31')]
                    
alb_sev_19 = alb_sev.loc[(df2['admission_date'] >= '2019-01-01')
                     & (df['admission_date'] <= '2019-12-31')]
                     
alb_sev_20 = alb_sev.loc[(df2['admission_date'] >= '2020-01-01')
                     & (df['admission_date'] <= '2020-12-31')]
```

```python
#Creating totals per race/type
albs_19 = pd.crosstab(alb_sev_19.race, alb_sev_19.severity)
albs_20 = pd.crosstab(alb_sev_20.race, alb_sev_20.severity)
nycs_19 = pd.crosstab(nyc_sev_19.race, nyc_sev_19.severity)
nycs_20 = pd.crosstab(nyc_sev_20.race, nyc_sev_20.severity)

#reseting index
albs_19 = albs_19.reset_index()
albs_20 = albs_20.reset_index()
nycs_19 = nycs_19.reset_index()
nycs_20 = nycs_20.reset_index()
```

```python
#Format albany dataframe to plot
frames = [albs_19, albs_20]
alb= pd.concat(
    frames,
    axis=0,
    join="outer",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)
alb.insert(1, "Race", ["Asian Pre", "Black Pre", "Unknown(Hispanic) Pre", "White Pre", 
                          "Asian Post", "Black Post", "Unknown(Hispanic) Post", "White Post"], True)
alb.drop(columns=['race']) 
            
#Format proper NYC dataframe to plot
frames = [nycs_19, nycs_20]
nyc= pd.concat(
    frames,
    axis=0,
    join="outer",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)
nyc.insert(1, "Race", ["Asian Pre", "Black Pre", "Other Pre", "White Pre", 
                          "Asian Post", "Black Post", "Other Post", "White Post"], True)
nyc.drop(columns=['race']) 
```

```python
sns.set_style("darkgrid")

#NYC plot
fig, ax = plt.subplots()
fig.suptitle('NYC VS. Albany Crimes By Race Pre/Post Bail Reform', size = 20)
ax = nyc.plot.bar(x= "Race", 
             rot = 45, 
             figsize=(12, 5),
             fontsize=12,
             stacked=True, 
             ax = ax, 
             )

vals = ax.get_xticks()
for tick in vals:
      ax.axvline(x=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1) 

#Title/Axis
ax.set_xlabel("Race", labelpad=20, size=15)  
ax.set_ylabel("NYC Arrests", labelpad=15, size=20)
ax.legend()

#Albany plot  
fig, ax2 = plt.subplots()
ax2 = alb.plot.bar(x= "Race", 
             rot = 45, 
             figsize=(12, 5),
             fontsize=12,
             stacked=True, 
             ax = ax2, 
             )
vals = ax2.get_xticks()
for tick in vals:
      ax2.axvline(x=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1) 
        
#Title/Axis
ax2.set_xlabel("Race", labelpad=20, size=15)  
ax2.set_ylabel("Albany Arrests", labelpad=15, size=20)
ax2.legend()
fig.savefig('full_figure.png')
```
